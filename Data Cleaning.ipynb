```python
import pandas as pd
import numpy as np

cahoots_2023 = pd.read_excel('call_data_from_CAHOOTS.xlsx')
cahoots_2021_2022 = pd.read_excel('call_data_from_CAHOOTS_2021_2022.xlsx')

cahoots = pd.concat([cahoots_2023, cahoots_2021_2022])
cahoots.drop(['Age', 'Gender', 'Race', 'Language','City','Reason for Dispatch'], axis=1, inplace=True)
cahoots_cleaned = cahoots.dropna(subset=['Date', 'TimeOfCall'])
cahoots_cleaned.tail(20)
```

```python
print(cahoots_cleaned.dtypes)
cahoots_cleaned.loc[:, "TimeOfCall"] = cahoots_cleaned["TimeOfCall"].astype(str)
cahoots_cleaned['Datetime'] = cahoots_cleaned['Date'].dt.strftime('%Y-%m-%d') + ' ' + cahoots_cleaned['TimeOfCall']
cahoots_cleaned.head()
```

```python
cahoots_cleaned.drop(['Date', 'TimeOfCall'], axis=1, inplace=True)
cahoots_cleaned.head()
```

```python
cahoots_cleaned['Datetime'] = pd.to_datetime(cahoots_cleaned['Datetime'], errors='coerce')
cahoots_cleaned = cahoots_cleaned.dropna(subset=['Datetime'])
cahoots_cleaned['Time'] = cahoots_cleaned['Datetime'].dt.floor('H')
df = cahoots_cleaned.groupby('Time').size().reset_index(name='Call Count')
df = df.sort_values(by="Time")
df = df.iloc[1:]
df.head()
```

```python
df.set_index('Time', inplace=True)
df = df.resample('H').asfreq().fillna(0).reset_index()
df['Call Count'] = df['Call Count'].astype(int)
df.head(20)
```

```python
header_file_path = 'headers.txt'  
data_file_path = 'CRNH0203-2023-OR_Corvallis_10_SSW.txt'

with open(header_file_path, 'r') as file:
    header_line = file.readlines()[1]
    headers = header_line.strip().split()

Weather2023 = pd.read_csv(data_file_path, delim_whitespace=True, names=headers)
Weather2023.head()
```

```python
data_file_path = 'Weather3.txt'

Weather2022 = pd.read_csv(data_file_path, delim_whitespace=True, names=headers)
Weather2022.head()
```

```python
data_file_path = 'Weather2021.txt'      

Weather2021 = pd.read_csv(data_file_path, delim_whitespace=True, names=headers)
Weather2021.head()
```

```python
weather_df = pd.concat([Weather2021, Weather2022, Weather2023])
weather_df.head(24)
```

```python
weather_df = weather_df.loc[:, ['LST_DATE', 'LST_TIME', 'T_HR_AVG','P_CALC']]
weather_df.rename(columns = {'LST_DATE':'Date'}, inplace = True)
weather_df.rename(columns = {'LST_TIME':'Hour'}, inplace = True)
weather_df.rename(columns = {'T_HR_AVG':'Temperature'}, inplace = True)
weather_df.rename(columns = {'P_CALC':'Perception'}, inplace = True)
weather_df.head(20)
```

```python
weather_df['Date'] = weather_df['Date'].astype(str)
weather_df['Hour'] = weather_df['Hour'].astype(str).str.zfill(4)  # Pad hours with zeros
weather_df['Time'] = pd.to_datetime(weather_df['Date'] + weather_df['Hour'], format='%Y%m%d%H%M')
weather_df = weather_df.drop(columns=['Date', 'Hour'])
weather_df.head()
```

```python
df = pd.merge(df, weather_df, on="Time", how="inner")
df = df.sort_values(by="Time")
df.head(20)
```

```python
# Adding the 'Date' column by extracting the date from the 'Time' column
df['Date'] = df['Time'].dt.date

# Adding the 'Hour' column by extracting the hour from the 'Time' column
df['Hour'] = df['Time'].dt.hour

# Adding the 'Day' column (Monday=1, Sunday=7)
df['Day'] = df['Time'].dt.dayofweek + 1

def get_season(date):
    month = date.month
    if month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    elif month in [9, 10, 11]:
        return 'Fall'
    else:
        return 'Winter'

df['Season'] = df['Date'].apply(get_season)
df.head(20)
```

```python
aqi_df1 = pd.read_csv("hourly_88101_2021.csv", low_memory=False)
aqi_df1 = aqi_df1[aqi_df1['County Name'] == 'Lane']
aqi_df2 = pd.read_csv("hourly_88101_2022.csv", low_memory=False)
aqi_df2 = aqi_df2[aqi_df2['County Name'] == 'Lane']
aqi_df3 = pd.read_csv("hourly_88101_2023.csv", low_memory=False)
aqi_df3 = aqi_df3[aqi_df3['County Name'] == 'Lane']
aqi_df = pd.concat([aqi_df1, aqi_df2, aqi_df3])
aqi_df.head()
```

```python
columns_to_drop = ['State Code', 'County Code', 'Parameter Code', 'POC',
                   'Latitude', 'Longitude', 'Datum', 'Parameter Name',
                   'Date GMT', 'Time GMT', 'Uncertainty', 'Qualifier', 'Method Type',
                   'Method Code', 'Method Name', 'State Name', 'County Name',
                   'Date of Last Change', 'MDL', "Units of Measure"]
aqi_df.drop(columns=columns_to_drop, inplace=True)
```

```python
unique_site_nums = aqi_df['Site Num'].unique()
print(unique_site_nums)
```

```python
site_nums_to_keep = [59, 60]
aqi_df = aqi_df[aqi_df['Site Num'].isin(site_nums_to_keep)]
```

```python
aqi_df['Datetime'] = pd.to_datetime(aqi_df['Date Local'] + ' ' + aqi_df['Time Local'], format='%Y-%m-%d %H:%M')
aqi_df.drop(columns=['Date Local', 'Time Local'], inplace=True)
aqi_df.head()
```

```python
aqi_df = aqi_df.groupby(['Datetime'], as_index=False)['Sample Measurement'].mean()
aqi_df.head()
```

```python
aqi_df = aqi_df.rename(columns={'Sample Measurement': 'AQI PM2.5', 'Datetime': 'Time'})
df = pd.merge(df, aqi_df, on="Time", how="inner")
df.head()
```

```python
df = df[(df['Temperature'] > -100) & (df['Temperature'] < 100)]  # Adjust the range as needed
df = df[(df['Perception'] > -1)]  # Adjust the range as needed
```
